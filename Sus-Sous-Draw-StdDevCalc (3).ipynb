```python
poseName = 'Sus-sous'
xMeanloaded = np.load('Stats/'+poseName+'MeanX.npy')
yMeanloaded = np.load('Stats/'+poseName+'MeanY.npy')
stdDevX = np.load('Stats/'+poseName+'StdDevX.npy')
stdDevY = np.load('Stats/'+poseName+'StdDevY.npy')
```

```python
#Calculate distance vectors
#1 zip the necessary lists - first Mean and stddevPlot Plus
xMeanloaded = np.load('Stats/'+poseName+'MeanX.npy')
yMeanloaded = np.load('Stats/'+poseName+'MeanY.npy')
MeanTuples = list(zip(xMeanloaded, yMeanloaded))
print(MeanTuples)

stdDevPlusX = np.load(os.path.join('Stats', poseName+'stDevPlotPlusX.npy'))
stdDevPlusY = np.load(os.path.join('Stats', poseName+'stDevPlotPlusY.npy'))
stdDevPlusTuples = list(zip(stdDevPlusX,stdDevPlusY))
print(stdDevPlusTuples)

longZip = list(zip(MeanTuples,stdDevPlusTuples))
print(longZip)
```

```python
import math


def calcDistance(x1,y1,x2,y2):
    distance = math.sqrt((x2-x1)**2 + (y2-y1)**2)
    return distance

def calcDistVector(longZipFile, filename):
    distanceVector = []
    for ((x1,y1),(x2,y2)) in longZipFile:
        d = calcDistance(x1,y1,x2,y2)
        print(d)
        distanceVector.append(d)
    np.save(os.path.join('Stats', filename), distanceVector)
```

```python
calcDistVector(longZip, poseName+'DistanceStdDevAdded.npy')#distance between Mean and standard dev added
```
