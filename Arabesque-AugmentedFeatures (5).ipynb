```python
#Helper Methods:
import math
def calcDistance(x1,y1,x2,y2):
    distance = math.sqrt((x2-x1)**2 + (y2-y1)**2)
    return distance

#def calcDistVector(longZipFile, filename):
    #distanceVector = []
    #for ((x1,y1),(x2,y2)) in longZipFile:
        #d = calcDistance(x1,y1,x2,y2)
        #print(d)
        #distanceVector.append(d)
    #np.save(os.path.join('Stats', filename), distanceVector)
```

```python
#Get the Mean Skeleton's coordinates
ArabesqueMeanX = np.load("Stats/ArabesqueMeanX.npy")
print(ArabesqueMeanX)
print(len(ArabesqueMeanX))
ArabesqueMeanY = np.load("Stats/ArabesqueMeanY.npy")
print(ArabesqueMeanY)
print(len(ArabesqueMeanY))
MeanTuples = list(zip(ArabesqueMeanX, ArabesqueMeanY))
print(MeanTuples)
```

```python
# loop through the ArabesqueCoords List.
#Loop Through each sample and calculate the distance 
#d between: (samplex1,sampley1) and (meanx1,meanx2)
#save the distance vectors per sample
POSE = "Arabesque"
allDistanceVectors = []
for sample in ArabesqueCoords:
    xCoords = sample[0::3]
    #print(xCoords)
    yPos = sample[1::3]
    yCoords = [-y for y in yPos] # need to make it negative to be consistent - the mean skeleton's y saved the negative coords.
    SampleTuples = list(zip(xCoords, yCoords))
    #print(SampleTuples)
    longZip = list(zip(MeanTuples, SampleTuples))
    #print(longZip)
    distanceVector = []
    for ((x1,y1),(x2,y2)) in longZip:
        d = calcDistance(x1,y1,x2,y2)
        #print(d)
        distanceVector.append(d)
    allDistanceVectors.append(distanceVector)

print(allDistanceVectors)
print(len(allDistanceVectors))
print(len(allDistanceVectors[0]))
np.save("DataForSVM/"+POSE+"DistanceVectorsPerSample.npy", allDistanceVectors)
    
```
