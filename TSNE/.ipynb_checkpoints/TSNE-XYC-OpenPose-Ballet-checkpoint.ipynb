```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.manifold import TSNE
import numpy as np
```

```python
X = np.load("../DataForSVM/XRootDATA.npy")
print(X[0])
print(len(X[0]))
print(len(X))
data = X
```

```python
y = np.load("../DataForSVM/yRootLabels.npy")
print(y[0])
print(len(y))
```

```python
for perplexity in [5, 30, 55, 80, 105,  130]:
    tsne = TSNE(n_components=2, random_state=0, perplexity=perplexity, learning_rate=10)
    tsne_out = tsne.fit_transform(data)
    tsne_frame = pd.DataFrame(np.row_stack( tsne_out ), columns=['x','y'])
    tsne_frame['label'] = y #baseline labels
    tsne_framesub = tsne_frame.copy()
    #tsne_framesub.label.cat.remove_unused_categories(inplace=True)
    sns.set_context("notebook",font_scale=1.1)
    sns.set_style("ticks")
    sns.lmplot(x='x', y='y',
               data=tsne_framesub,
               fit_reg=False, legend=True,
               height=8,
               hue='label',
               scatter_kws={"s":50, "alpha":0.2})
```
